```python
import pandas as pd
import matplotlib.pyplot as plt
```

LOAD

```python
df = pd.read_csv('Airline_Delay_Cause.csv')
df.head()
```

```python
delay_cols = [
    "carrier_delay",
    "weather_delay",
    "nas_delay",
    "security_delay",
    "late_aircraft_delay"
]

total_delays = df[delay_cols].sum()

total_delays.plot(kind="bar")
plt.title("Total Delay Time by Cause")
plt.ylabel("Minutes of Delay")
plt.xlabel("Delay Cause")
plt.show()
```
